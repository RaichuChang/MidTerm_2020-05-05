```
import thinkdsp as td

x= td.UncorrelatedGaussianNoise()
x= x.make_wave(duration= 1, framerate= 10000)
x
```

```
x.ts, x.ys
x.ys.shape
```

```
import matplotlib.pyplot as pl
pl.plot(x.ts, x.ys)
```

```
x.ys.max(), x.ys.mean(), x.ys.std()
```

```
sx= x.make_spectrum()
sx.fs, sx.hs 
sx.hs.shape
```

```
sx.plot()
```

```
sx.plot_power()
```

```
Sx= sx.make_integrated_spectrum()
Sx.cs.shape
```

```
Sx.plot_power()
```

```
Sx.estimate_slope( )
```

```
Sx.plot_power()
pl.xscale('log')
pl.yscale('log')
pl.grid(True)
```

```
Sx.estimate_slope()
```

```
y= x.cumsum()
x.ys.shape, y.ys.shape
```

```
y.plot()
```

```
y.ys.max(), y.ys.mean(), y.ys.std()
```

```
sy= y.make_spectrum()
sy.plot_power()
pl.xscale('log')
pl.yscale('log')
pl.grid(True)
```

```
sy.make_integrated_spectrum().plot_power()
pl.xscale('log')
pl.yscale('log')
pl.grid(True)
```

```
sy.make_integrated_spectrum().estimate_slope()
```
